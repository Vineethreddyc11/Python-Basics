# IF,ELSE,ELIF

- if…elif…else are Conditional Statements that provide you with the decision making that is required when you want to execute code based on a particular condition.
      
- The if…elif…else statement used in Python helps automate that decision making process.
      
- While writing code in any language, you will have to control the flow of your program. This is generally the case when there is decision making involved - you will want to execute a certain line of codes if a condition is satisfied, and a different set of code incase it is not. In Python, you have the if, elif and the else statements for this purpose. 
      

### Indentation

- Python relies on indentation (whitespace at the beginning of a line) to define scope in the code. Other programming languages often use curly-brackets for this purpose.

# if condition

- The if condition is considered the simplest of the three.
- if statement makes a decision based on whether the condition is true or not. 
- If the condition is true, it prints out the indented expression.
- If the condition is false, it skips printing the indented expression.

- 

-   

#### Syntax



```python
if condition:
    indented statement block 
```

##### Examples

```python
a = 15

if a>10:
    print("a is greater")
```

```python
if 10<15:
    print("10 is smaller than 15")
```

```python
score_theory = 40
score_practical = 45
if score_theory + score_practical > 100: # End of condition
    print("Invalid score. Please check the input.")
```

```python
score_theory = 55
score_practical = 50
if score_theory + score_practical > 100: # End of condition
    print("Invalid score. Please check the input.")
```

# else condition

- The if-else condition adds an additional step in the decision-making process compared to the simple if statement. 
- The beginning of an if-else statement operates similar to a simple if statement; 
- However, if the condition is false, instead of printing nothing, the indented expression under else will be printed.



- 

- 

#### Syntax

```python
if condition:
      Indented statement block for when condition is TRUE
else:
      Indented statement block for when condition is FALSE
```

##### Examples

```python
if 1>2:
    print("Its True")
else:
    print("False")
```

```python
age = float(input("Enter your age:"))
if age>=18:
    print("You are eligible voting")
else:
    print("Not eligible for voting")
```

```python
mat_mark = 45
phy_mark = 40
total = mat_mark+phy_mark
if total>100:
    print('error')
else:
    print(total)
```

```python
length = int(input('Enter Lenght'))

breadth = int(input('Enter Breadth'))
if length == breadth:
    print("Its Square")
else:
    print("Its Rectangle")
```

```python
first_num = int(input('Enter 1st number:'))
second_num = int(input('Enter 2nd number:'))
if first_num>second_num:
    print("Greatest among 2 numbers is:", first_num)
else:
    print("Greatest among 2 numbers is:",second_num)
```

- A student will not be allowed to sit in exam if his/her attendence is less than 75%.
Take following input from user
Number of classes held
Number of classes attended.
And print
percentage of class attended
Is student is allowed to sit in exam or not.

```python
Name  = str(input("Enter Name:"))
classes_held  = int(input("No.of Classes held:"))
classes_attended = int(input("No.of classes attended:"))
percentage_of_attendence = classes_attended/classes_held*100

if percentage_of_attendence >= 75:
    print(percentage_of_attendence,'Percent')
    print(Name,"is allowed to sit in exam")
else:
    print(percentage_of_attendence)
    print(Name,"is not allowed to sit in exam")
```

# elif condition

- The most complex of these conditions is the if-elif-else condition. 
- When you run into a situation where you have several conditions, you can place as many elif conditions as necessary between the if condition and the else condition.

- 

- 

#### Syntax

```python
if Condition1:
      Indented statement block for Condition1
elif Condition2:
       Indented statement block for Condition2
else:
       Alternate statement block if all condition check above fails
```

###### Examples

```python
bill  = int(input("Enter Price:"))
if bill>1000:
    print("Discount available of 10%")
    discount = bill*10/100
    print("Discount amount:",discount)
    amt = bill-discount
    print("Amount payable:",amt)
elif 500<bill<1000:
    print("Discount available of 5%")
    discount = bill*5/100
    print("Dicount amount:",discount)
    amt = bill-discount
    print("Amount payable:",amt)
else:
    print("Amount payable:",bill)
```

- Given an integer,n , perform the following conditional actions:
If n is odd, print Weird
If n is even and in the inclusive range of 2 to 5, print Not Weird
If n is even and in the inclusive range of  6 to 20, print Weird
If n is even and greater than 20, print Not Weird

```python
number = int(input("Enter Number:"))

if number%2 ==1 :
    print("Weird")
elif number%2 == 0 and 2<=number<=5 :
    print("Not Weird")
elif number%2 == 0 and 6<=number<=20 :
    print("Weird")
else:
    print("Not Weird")
```

- A school has following rules for grading system:
a. Below 25 - F
b. 25 to 45 - E
c. 45 to 50 - D
d. 50 to 60 - C
e. 60 to 80 - B
f. Above 80 - A
Ask user to enter marks and print the corresponding grade.

```python
marks = int(input("Enter Marks:"))

if marks < 25:
    print("F")
elif 25<marks<40:
    print("E")
elif 45<marks<50:
    print("D")
elif 50<marks<60:
    print("C")
elif 60<marks<80:
    print("B")
else:
    print("A")
```
